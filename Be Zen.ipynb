```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading the CSV

```python
df=pd.read_csv('2022_02_08-02_30_31_AM.csv')
```

```python
df.head(15)
```

# Products without prices

```python
df_withoutPrices=df[df['price_string'].isna()]
```

```python
df_withoutPrices.reset_index(inplace=True)
```

```python
df_withoutPrices
```

# Products with Prices

```python
df_prices=df[df['price_string'].notna()]
```

```python
df_prices.reset_index(inplace=True,drop=True)
```

```python
df_prices
```

# To remove $ sign

```python
df_nodollar=df_prices
```

```python
df_nodollar['price_string']=df_prices['price_string'].str.strip('$').astype(float)
```

# To add $ sign

```python
df_dollar=df_prices
```

```python
df_dollar['price_string']= '$' + df_nodollar['price_string'].astype(str)
```

```python
df_dollar
```

```python
df_dollar[['Currency', 'Value']] = df_dollar['price_string'].str.split('$', expand=True)
```

```python
df_dollar['Currency']='$'
```

```python
df_dollar['Value']=df_dollar['Value'].astype(float)
```
